```python
ratings = pd.merge(ratings, movies)
ratings
```

```python
ratings.title.value_counts()
```

```python
ratings.movie_id.nunique()
```

```python
movie_grouped = ratings.groupby('title').agg({'rating': [np.size,np.sum, np.mean]})
movie_grouped.head()
```

```python
popular_movies = movie_grouped['rating']['size'] >= 100
popular_movies.head() ,popular_movies.shape
```

```python
popular_movies= movie_grouped[popular_movies].sort_values([('rating', 'mean')], ascending=False)
popular_movies.head(), popular_movies.shape
```

```python
grouped_sum = movie_grouped['rating']['size'].sum()
```

```python
popular_movies['percentage'] = popular_movies['rating']['size'].div(grouped_sum)*100
popular_movies.head(), popular_movies.shape
```

```python
popular_movies = popular_movies.sort_values(('percentage'), ascending=False)
popular_movies, popular_movies.shape
```

```python
popular_movies["Rank"] = popular_movies["percentage"].rank(ascending=False)
```
